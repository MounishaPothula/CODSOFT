Import Required Libraries

```python
# Importing necessary libraries for data handling, visualization, and warnings suppression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Suppress warnings for a cleaner output
warnings.simplefilter(action='ignore', category=FutureWarning)

# Enable inline plotting for better visualization in notebooks
%matplotlib inline
```

```python
# Load the dataset into a Pandas DataFrame
df = pd.read_csv("creditcard.csv")
```

```python
# Display the first five rows of the dataset to get an overview
df.head()
```

```python
# Get the number of rows and columns in the dataset
df.shape
```

```python
# Generate descriptive statistics for numerical features and transpose the output for better readability
df.describe().T
```

```python
# Display concise summary of the dataset, including data types and non-null counts
df.info()
```

```python
# Check for missing values by summing up null values in each column
df.isnull().sum()
```

```python
# Removes rows with any NaN values
df = df.dropna()
```

Exploratory Data Analysis (EDA) & Visualization

```python
# Visualizing the class distribution to check the imbalance in fraudulent vs. non-fraudulent transactions
sns.countplot(x=df["Class"])
```

```python
# Drop 'Time', 'Amount', and 'Class' columns to focus on feature distributions
df_temp = df.drop(columns=["Time", "Amount", "Class"], axis=1)

# Create distribution plots for each feature
fig, ax = plt.subplots(nrows=7, ncols=4, figsize=(20, 40))  # Define subplot grid (7x4)
ax = ax.flatten()  # Flatten the axes array for easier iteration

# Plot the distribution of each feature
for index, col in enumerate(df_temp.columns):
    sns.histplot(df_temp[col], ax=ax[index], kde=True)  # Changed from distplot to histplot (latest Seaborn version)

plt.tight_layout(pad=0.5, w_pad=0.5, h_pad=5)  # Adjust layout for better spacing
```

```python
# Visualizing the distribution of the 'Time' feature
sns.histplot(df["Time"], kde=True)
```

```python
# Visualizing the distribution of the 'Amount' feature
sns.histplot(df["Amount"], kde=True)
```

```python
# Separating features (X) and target variable (y)
X = df.drop(columns=["Class"], axis=1)  # Features (all columns except 'Class')
y = df["Class"]  # Target variable (Fraud or Non-Fraud)
```

Feature Scaling Using Standardization

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_scaler = sc.fit_transform(X)
```

```python
X_scaler
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.25, random_state=42, stratify=y)
```

Training Machine Learning Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
print("F1 Score: ", f1_score(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, f1_score
model = RandomForestClassifier()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
print("F1 Score: ", f1_score(y_test, y_pred))
```

```python
# Visualizing class distribution in the training set
sns.countplot(x=y_train)
plt.title("Class Distribution in Training Data")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

Handling Class Imbalance with SMOTE

```python
from imblearn.over_sampling import SMOTE
over_sample = SMOTE()
X_smote,y_smote = over_sample.fit_resample(X_train, y_train)
```

```python
# Visualizing class distribution after applying SMOTE
sns.countplot(x=y_smote)
plt.title("Class Distribution After SMOTE Oversampling")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score

model = LogisticRegression()
model.fit(X_smote, y_smote)

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
print("F1 Score: ", f1_score(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, f1_score
model = RandomForestClassifier()
model.fit(X_smote, y_smote)

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
print("F1 Score: ", f1_score(y_test, y_pred))
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, f1_score
model = XGBClassifier()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
print("F1 Score: ", f1_score(y_test, y_pred))
```
